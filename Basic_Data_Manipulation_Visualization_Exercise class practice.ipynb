### 49. Write a Python script to display the

- c. Month of year
- d. Week number of the year
- e. Weekday of the week
- f. Day of year
- g. Day of the month
- h. Day of week

```python
today_dt = dt.datetime.now()
print('current date time is ',today_dt)
print('current year is ', today_dt.year)
print('curent month is ', today_dt.month)
print('week number of year is ', today_dt.isocalendar().week)
print('week day of the week is ', today_dt.strftime('%A'))
print('day of the year ',today_dt.strftime('%j'))
print('day of the month is ', today_dt.strftime('%d'))
print('day of week ', today_dt.weekday())
```
